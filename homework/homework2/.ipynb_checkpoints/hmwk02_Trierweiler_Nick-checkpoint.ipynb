In soccer tournaments, [penalty shootouts](https://en.wikipedia.org/wiki/Penalty_shoot-out_(association_football) are used to determine the outcome of the game if the score is tied after regulation play.   In a penalty shootout, each team takes turns attempting 5 shots on the goal from the penalty mark that are defended only by the opposing team's goalkeeper, with the team that makes more successful kicks being declared the winner of the match.  If after 5 shots the score is still tied, the game enters _sudden death_ and the teams continue to take 1 shot each until one team scores and the other does not. 

Suppose that two teams ( Team $A$ and Team $B~$) are still tied at the end of the first round of 5 penalty shots and are entering sudden death.  Each team will now take one shot each in an attempt to win.  Suppose that the result of one shot is independent of the other shots, Team $A$ makes their shot with probability $p$, and Team $B$ makes their shot with probability $q$.  Let $W$ be the event that "Team $A$ wins the game", $E$ be the event that "the game ends after the first round of sudden death penalty shots", and $T$ be the event that "the game is still tied after the first round of sudden death penalty shots". 

**Part A**: The shootout continues until Team $A$ or Team $B$ wins.  What is the sample space $\Omega$ for this experiment?  Represent the outcomes of individual shots by $G$ for a goal and $M$ for a miss and assume that Team $A$ shoots before Team $B$.

Sample Space: $\Omega = \{GG, GM, MG, MM\}$

Probability that A scores = $p$

Probability that B scores = $q$


1. $W$ = {GM} # event Team  A  wins the game 
2. $E$ = {GM, MG} # event the game ends after the first round of sudden death penalty shots
3. $T$ = {GG, MM} # event the game is still tied after the first round of sudden death penalty shots
4. $L$ = {MG} # event Team B wins, A loses


**Part B**: Determine $P(W \mid E)$ in terms of $p$ and $q$.

$P(W \mid E) = \frac{P(E \mid W)P(W)}{P(E)}$

$P(W \mid E) = \frac{P(E \mid W)P(W)}{P(E \mid W)P(W) + P(E \mid L)P(L) + P(E \mid T)P(T)}$

$P(W \mid E) = \frac{1 \times p(1-q)}{1 \times q(1-p) + 1 \times p(1-q) + 0}$ P(E | T)P(T) is 0 in the event of a win

$P(W \mid E) = \frac{p(1-q)}{q + p - 2pq} = \frac{p-pq}{q + p - 2pq}$

**Part C**: Explain why events $W$ and $T$ are independent.

$P(W) = \frac{p(1-q)}{p+q-2qp}$

```python
R = {"name": "R", "loc": np.array(["H", "M", "K", "S"]), 'probs' : np.array([1/4,1/4,1/4,1/4])}
H = {"name": "H", "loc": np.array(["R", "M"]), 'probs' : np.array([1/2,1/2])}
M = {"name": "M", "loc": np.array(["R", "H", "K"]), 'probs' : np.array([1/3,1/3,1/3])}
K = {"name": "K", "loc": np.array(["R", "M", "S", "E"]), 'probs' : np.array([1/4,1/4,1/4,1/4])}
S = {"name": "S", "loc": np.array(["R", "K", "E"]), 'probs' : np.array([1/3,1/3,1/3])}
E = {"name": "E", "loc": np.array(["K", "S"]), 'probs' : np.array([1/2,1/2])}
Jerry = {"loc": np.array([H, M, K, S]), 'probs' : np.array([1/4,1/4,1/4,1/4])}
Steve = {"loc": np.array([K, S]), 'probs' : np.array([1/2,1/2])}
PartA = {"loc": np.array([H, K, R]), 'probs' : np.array([1/3,1/3,1/3])}
PartB = {"loc": np.array([H, M, K, S]), 'probs' : np.array([1/4,1/4,1/4,1/4])}



def sample_A():
    first = np.random.choice(PartA['loc'] ,p = PartA['probs'])
    second = np.random.choice(first['loc'] ,p = first['probs'])
    if(second == "R"):
        return(True)
    else:
        return(False)
    
def sample_B():
    first = np.random.choice(PartB['loc'] ,p = PartB['probs'])
    second = np.random.choice(first['loc'] ,p = first['probs'])
    if((second == "K") or (first['name'] == "K")):
        return(True)
    else:
        return(False)
    
def sample_first(J, S):
    J = np.random.choice(J['loc'] ,p = J['probs'])
    S = np.random.choice(S['loc'] ,p = S['probs'])
    if(S == J):
        return True
    else: 
        return False

def sample_second(J, S):
    J = np.random.choice(J['loc'] ,p = J['probs'])
    S = np.random.choice(S['loc'] ,p = S['probs'])
    return(sample_first(J,S))



#sample first, but requires some conversions after for np.sum() so I wanted to keep this separate for cleanliness
def sample_both(J, S): 
    J = np.random.choice(J['loc'] ,p = J['probs'])
    S = np.random.choice(S['loc'] ,p = S['probs'])
    if(S == J):
        return J
    else: 
        return ""

    
def match_or_not(selection, num_samples = 10000):
    if(selection == "first"):
        arr = np.array([sample_first(R, E) for i in range(num_samples)])
        return np.sum(arr / num_samples)
    elif(selection == "second"):
        arr = np.array([sample_second(Jerry, Steve) for i in range(num_samples)])
        return np.sum(arr / num_samples)
    elif(selection == "both"):
        #
        arr = np.array([sample_both(R, E) for i in range(num_samples)])
        arr2 = np.array([sample_both(R, E) if(arr[ii]!="") else "" for ii in range(num_samples)])
        arr3 = np.array([True if(arr2[ii] != "") else False for ii in range(num_samples)])
        return(np.sum(arr3 / num_samples))
    elif(selection == "A"):
        arr = np.array([sample_A() for i in range(num_samples)])
        return(np.sum(arr / num_samples))
    elif(selection == "B"):
        arr = np.array([sample_B() for i in range(num_samples)])
        return(np.sum(arr / num_samples))
    else:
        return("error")
    
samples = 100000

print("number of samples used: {} (10000 is default)".format(samples))
#PartA
A_ans = match_or_not("A", samples)    
print("Part A:")
print("{:.2f}% chance Steve ends up at R from M".format(A_ans*100))  

#PART B
B_ans = match_or_not("B", samples)    
print("Part B:")
print("{:.2f}% chance Steve goes to R first or second".format(B_ans*100))


#PART C

FIRST = match_or_not("first", samples)
SECOND = match_or_not("second", samples)
BOTH = match_or_not("both", samples)
OR = FIRST + SECOND - BOTH


print("PART C:")
print("{:.2f}% Jerry and Steve match first locations".format(FIRST*100)) #first choice match: 0.250081 with 1 million samples
print("{:.2f}% Jerry and Steve match second locations".format(SECOND*100)) #second choice match: 0.204286 with 1 million samples
print("{:.2f}% Jerry and Steve match first OR second locations".format(OR*100))
print("FOR MAXIMUM FUN {:.2f}% Jerry and Steve match first AND second locations".format(BOTH*100))
print("DONE! :D")

```

In this problem you will write some code to simulate the game of **Miniopoly** (a very rough approximation of [Monopoly](https://en.wikipedia.org/wiki/Monopoly_(game)).  The game consists of a round board with 40 "squares" numbered from $0$ to $40$, where squares $0$ and $40$ are [coincident](https://en.wikipedia.org/wiki/Coincident) and called **GO**.  A player starts on **GO** with \$200 in their pocket.  On each turn the player rolls a pair of dice and then moves a number of spaces equal to the sum of the dice.  Landing on different squares have different results: 

- Squares 2, 7, 17, 22, 33, and 36 cause you to draw a card.  The cards tell you to collect \$50 or \$100, or pay the bank \$100, \$125, or \$150.  

<p> </p>

- Squares 0, 10, and 20 do not trigger any action 

<p> </p>

- Square 30 is **Go To Jail**, which sends the player directly to **Jail** and ends the turn 

<p> </p>

- If a player starts a turn from **Jail** she rolls two dice: 

    - If the player rolls doubles (i.e. the dice have equal values) the player moves from Square 10 a number of squares equal to the sum of the dice
    
    - If the player rolls anything other than doubles she pays the bank \$10 and remains in **Jail** 
    
<p> </p>

- If the player passes or lands on **GO** she collects \$200. 

<p> </p>

- If the player lands on any other square not mentioned above, she pays the bank an amount equal to the square's number (e.g. if she lands on Square 29 she pays the bank \$29). 

<p> </p>

- If the player rolls doubles, she gains $10\times$ the doubled number, in addition to whatever other actions her roll may have caused. For example, rolling two sixes results in a payment of \$60. This rule does not apply if the player begins the turn from Jail.

<p> </p>

If at any point the player has \$0 or less they go bankrupt and the game ends. 

<br>

**Part A**: Complete the function miniopoly_turn below. The function should simulate one turn of Miniopoly.  It should take as arguments the player's current location on the board as well as the player's current cash reserves, and return the player's new location and cash reserves after the turn has completed. Note that if a player goes bankrupt during the game you should return their cash as 0 instead of a negative value.  

```python
#python dictionaries for probability 
dice = {'roll' : np.array([1,2,3,4,5,6]), 'probs' : np.array([1/6,1/6,1/6,1/6,1/6,1/6])}
deck = {'card' : np.array([50,100,-100,-125,-150]), 'probs' : np.array([1/5,1/5,1/5,1/5,1/5])}

#state arrays
draw = [2,7,17,22,33,36]
no_action = [0,10,20]
jail = [30]

#default vars
cash_start = 200
state_start = 0

def miniopoly_turn(state, cash, jailed):
    if(cash > 0):
        d1 = np.random.choice(dice['roll'], p = dice['probs'])
        d2 = np.random.choice(dice['roll'], p = dice['probs'])
        doubles = (d1 == d2)
        if(state == 'jail'):
            if(doubles):
                state = 10 #movement is below in the case of escape...
                cash -= d1 * 10 #negate double cash since started in jail
            else:
                cash -=10
        if(state != 'jail'): #state is a integer value, not 'jail'
            if(doubles):
                cash += d1 * 10
            state += (d1+d2) #player moves forward
            #check if they pass go!
            if(state >= 40):
                state = state%40
                cash +=200
            #NOW CHECK STATE OF THE NEW LOCATION
            if(state in draw):
                cash += np.random.choice(deck['card'], p = deck['probs'])
            elif(state in no_action):
                state = state
                cash = cash
            elif(state in jail):
                state = 'jail'
                jailed = True
            else:
                cash -= state
            return(state, cash, jailed)
    else:
        return(state, 0, jailed)
    
    return(state, cash, jailed)
```

**Part B**: Use your function from **Part A** to Simulate at least 5000 30-turn games of Miniopoly and make a frequency histogram of the player's cash at the end of the game. 

```python
def thirty_turn_game(state, cash, jailed, num_turns = 30):
    for i in range(num_turns):
        state, cash, jailed = miniopoly_turn(state, cash, jailed)
    return(state, cash, jailed)
    
def sample(state, cash, jailed, num_samples):
    cash_jailed_arr = np.array([thirty_turn_game(state, cash, jailed) for ii in range(num_samples)])
    #print(cash_jailed_arr)
    return cash_jailed_arr

game_samples = 100000

results_cash_arr = sample(0, 200, False, game_samples)
#store dataframe for use in part C :D
df = pd.DataFrame({"state": (results_cash_arr[:,0]), "cash": results_cash_arr[:,1], "jailed": results_cash_arr[:,2]}, dtype=int)
df['jailed'] = (df['jailed'] == "True") #convert to bool, and dont lose my values thanks!
#print(df['jailed'])
cash = np.array(df['cash'])

fig = plt.figure()
ax = fig.add_subplot(111)
n, bins, patches = ax.hist(cash, weights=np.zeros_like(cash) + 1. / cash.size)
ax.set_xlabel('Cash', size=20)
ax.set_ylabel('Percentage of Players', size=20)
ax.legend
plt.show()
df.dtypes

```
